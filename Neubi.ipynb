```python
import datetime
from matplotlib import pyplot
start = datetime.datetime.now().timestamp()
a = 0
y = []
n = int(input("Int : "))
print(">> n * n : ", n * n)
for i in range(n):
    a += n
    y.append(datetime.datetime.now().timestamp() - start)
    
pyplot.plot(list(range(n)), y) 
pyplot.xlabel("Loop N")
pyplot.ylabel("Time (s)")
pyplot.show()

```
